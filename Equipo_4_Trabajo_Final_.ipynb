Contamos con las siguientes variables:

1. ID: ID de solicitud de préstamo del cliente
2. Client_Income: Ingresos del cliente en \$
3. Car_Owned: Cualquier automóvil que poseía el cliente antes de solicitar el préstamo para otro automóvil (0 significa No y 1 significa lo contrario)
4. Bike_Owned: Cualquier bicicleta que poseía el cliente (0 significa No y 1 significa lo contrario)
5. Active_Loan: Cualquier otro préstamo activo en el momento de la solicitud del préstamo (0 significa No y 1 significa lo contrario)
6. House_Own: Cualquier casa que poseía el cliente (0 significa No y 1 significa lo contrario)
7. Child_Count: Número de hijos que tiene el cliente
8. Credit_Amount: Monto del crédito del préstamo en \$
9. Loan_Annuity: Anualidad del préstamo en \$
10. Accompany_Client: Quién acompañó al cliente cuando solicitó el préstamo
11. Client_Income_Type: Tipo de ingresos del cliente
12. Client_Education: Nivel más alto de educación alcanzado por el cliente
13. Client_Marital_Status: Estado civil del cliente (D- Divorciado, S- Soltero, M- Casado, W- Viudo)
14. Client_Gender: Género del cliente
15. Loan_Contract_Type: Tipo de préstamo (CL- Préstamo en efectivo, RL- Préstamo renovable)
16. Client_Housing_Type: Situación de vivienda del cliente
17. Population_Region_Relative: Población relativa de la región donde vive el cliente. Un valor más alto significa que el cliente vive en un área más poblada
18. Age_Days: Edad del cliente al momento de enviar la solicitud
19. Employed_Days: Días antes de la solicitud, el cliente comenzó a ganar
20. Registration_Days: Días antes de la solicitud del préstamo, el cliente cambió su registro
21. ID_Days: Días antes de la solicitud del préstamo, el cliente cambió su documento de identidad con el que solicitó el préstamo
22. Own_House_Age: Antigüedad de la casa del cliente en años
23. Mobile_Tag: Número de móvil proporcionado por el cliente (1 significa Sí y 0 significa No)
24. Homephone_Tag: Número de teléfono de casa proporcionado por el cliente (1 significa Sí y 0 significa No)
25. Workphone_Working: ¿Se podía acceder al número de teléfono del trabajo? (1 significa Sí y 0 significa No)
26. Client_Occupation: Tipo de ocupación del cliente
27. Client_Family_Members: Número de familiares que tiene el cliente
28. Cleint_City_Rating: calificación de la ciudad del cliente. 3 indica la mejor, 2 indica buena y 1 indica promedio
29. Application_Process_Day: día de la semana en el que el cliente solicitó el préstamo (0-domingo, 1-lunes, 2-martes, 3-miércoles, 4-jueves, 5-viernes, 6-sábado)
30. Application_Process_Hour: hora del día en el que el cliente solicitó el préstamo
31. Client_Permanent_Match_Tag: indicación si la dirección de contacto del cliente no coincide con la dirección permanente.
32. Client_Contact_Work_Tag: indicación si la dirección de trabajo del cliente no coincide con la dirección de contacto.
33. Type_Organization: tipo de organización en la que trabaja el cliente
34. Score_Source_1: puntuación obtenida de otra fuente. Esta es una fuente normalizada
35. Score_Source_2: puntuación obtenida de otra fuente. Esta es una fuente normalizada
36. Score_Source_3: Puntuación obtenida de otra fuente. Esta es una fuente normalizada
37. Social_Circle_Default: Cuántos amigos o familiares del cliente no cumplieron con algún pago del préstamo en los últimos 60 días
38. Phone_Change: Cuántos días antes de la solicitud del préstamo, el cliente cambió su teléfono
39. Credit_Bureau: Número total de consultas en el último año
40. Default: 1 significa que el cliente no cumplió con los pagos del préstamo y 0 significa lo contrario




```python
# Importamos Librerías
import pandas as pd                 # Biblioteca para la manipulación y análisis de datos mediante DataFrames
import seaborn as sns               # Biblioteca para la creación de gráficos estadísticos con diseño personalizable
import numpy as np                  # Biblioteca para operaciones matemáticas y manejo de datos numéricos
import matplotlib.pyplot as plt     # Módulo de Matplotlib para la creación de gráficos personalizados
```

```python
# Definición de la variable "datos_0", que contiene la URL en formato RAW del conjunto de datos alojado en GitHub
#Ya que en el conjunto de datos "Train_Dataste" se hará la construcción del modelo
datos_0 = "https://raw.githubusercontent.com/DavidDiaz24/TSC/refs/heads/main/Train_Dataset.csv"
```

```python
# Creación del DataFrame "datos_entrenar" a partir del archivo .csv con pandas
datos_entrenar = pd.DataFrame(pd.read_csv(datos_0))
```

```python
# Configuración de Pandas para mostrar todas las columnas en la salida
#y que Pandas no trunque la cantidad de columnas mostradas ya que hay muchas
pd.set_option('display.max_columns', None)

# Mostramos las primeras 5 filas del DataFrame
datos_entrenar.head()
```

```python
#Mostramos las dimensiones (filas y columnas) del DataFrame
datos_entrenar.shape
```

```python
# Eliminar columnas irrelevantes para el análisis: "ID", "Accompany_Client", "Application_Process_Day", "Application_Process_Hour"
# Además, se eliminan columnas con bajo valor informativo o que son redundantes: "Client_Occupation", 'Type_Organization', "Mobile_Tag"
datos_entrenar.drop(["ID", "Accompany_Client", "Application_Process_Day", "Application_Process_Hour", "Client_Occupation", 'Type_Organization', "Mobile_Tag"], axis=1, inplace=True)
datos_entrenar.head()
```

```python
# Extraemos los nombres de las columnas para identificar las variables disponibles en nuestro conjunto de datos
# Los convertimos en una lista para analizar y manipular fácilmente las variables
columnas = datos_entrenar.columns
columnas = list(columnas)
```

```python
# Mostramos un resumen del DataFrame para analizar la estructura de los datos, tipos de variables, y valores no nulos
datos_entrenar.info()
```

```python
# Creamos una lista que contiene variables numéricas relevantes para el análisis
# Estas variables incluyen información financiera, demográfica y de cómo se comporta el cliente
varibales_numericas = ['Client_Income',
 'Car_Owned',
 'Bike_Owned',
 'Active_Loan',
 'House_Own',
 'Child_Count',
 'Credit_Amount',
 'Loan_Annuity',
 'Population_Region_Relative',
 'Age_Days',
 'Employed_Days',
 'Registration_Days',
 'ID_Days',
 'Own_House_Age',
 'Homephone_Tag',
 'Workphone_Working',
 'Client_Family_Members',
 'Cleint_City_Rating',
 'Score_Source_1',
 'Score_Source_2',
 'Score_Source_3',
 'Social_Circle_Default',
 'Phone_Change',
 'Credit_Bureau',
 'Default']
```

```python
# Convertir los valores de las variables numéricas a tipo numérico si no lo son
# El argumento 'errors="coerce"' convierte valores no válidos como texto, nulos, etc. en NaN, para así asegurar que las columnas sean consistentes
for i in varibales_numericas:
  datos_entrenar[i] = pd.to_numeric(datos_entrenar[i], errors='coerce')
```

```python
# Verificamos nuevamente la estructura del DataFrame después de convertir las variables numéricas
# Para asegurar que los cambios realizados se hayan aplicado correctamente, por tro lado tambien observamos que hay varias columnas con valores faltantes
datos_entrenar.info()
```

```python
variables_categoricos = []
for i in columnas:
  if i not in varibales_numericas:
    variables_categoricos.append(i)
for i in ['Car_Owned', 'Bike_Owned', 'Active_Loan', 'House_Own', 'Child_Count', 'Homephone_Tag', 'Workphone_Working', 'Client_Family_Members', 'Cleint_City_Rating', 'Credit_Bureau', 'Default']:
  variables_categoricos.append(i)
variables_categoricos = set(variables_categoricos)
variables_categoricos = list(variables_categoricos)
variables_categoricos
```

```python
for i in variables_categoricos:
  if i in varibales_numericas:
    varibales_numericas.remove(i)
varibales_numericas
```

```python
# Saber cuantas variables dummies tendriamos por cada variable categorica
for i in variables_categoricos:
  print(datos_entrenar[i].value_counts())
  print("-" * 85)
```

```python
variables_cat_dummies = ["Client_Income_Type", "Client_Education", "Client_Marital_Status", "Client_Gender", "Loan_Contract_Type", "Client_Housing_Type", "Client_Permanent_Match_Tag", "Client_Contact_Work_Tag"]
# Con drop_first evitamos colinealidad
for i in variables_cat_dummies:
  i = pd.get_dummies(datos_entrenar[i], dtype = int, drop_first = True)
  datos_entrenar = pd.concat([datos_entrenar, i], axis=1)
```

```python
for i in variables_cat_dummies:
  datos_entrenar.drop(i, axis=1, inplace=True)
```
