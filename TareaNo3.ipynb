```python
#Se importa la libreria Pandas como pd
import pandas as pd
#Se importa la funcion drive
from google.colab import drive
#Conectamos el contenido de nuestro Google Drive con el Google Colab
drive.mount('/content/drive')
```

```python
#Agregamos la ubicacion del archivo "tabla1" y lo nombramos como "tabla_1"
tabla_1 = "/content/drive/MyDrive/Temas Selectos/Tarea No 3/tabla1.xlsx"
#Leemos el excel que nombramos en la linea de arriba
datos_1 = pd.read_excel(tabla_1)
#Mostramos los primeros 5 datos de la tabla_1
datos_1.head()
```

```python
#Agregamos la ubicacion del archivo "tabla1" y lo nombramos como "tabla_1"
tabla_2 = "/content/drive/MyDrive/Temas Selectos/Tarea No 3/tabla2.xlsx"
#Leemos el excel que nombramos en la linea de arriba
datos_2 = pd.read_excel(tabla_2)
#Mostramos los primeros 5 datos de la tabla_2
datos_2.head()
```

```python
# Combinamos los 2 exceles
datos_combinados = pd.merge(datos_1, datos_2, left_on='fips', right_on='county_fips')
#Mostramos los primeros 5 datos de la combinada
datos_combinados.head()
```

```python
#Eliminanos las columnas repetidos o con valores reduntantes
#axis=1: Indica que queremos eliminar una columna (axis=0 sería para filas).
datos_combinados = datos_combinados.drop([ 'state','county_y','county_fips','county_ascii'], axis=1)
#Mostramos los primeros 5 datos de la combinada
datos_combinados.head()
```

```python
# Agregar la columna 'year' con el valor 2023
datos_combinados.insert(1, 'year', 2023)
#Mostramos los primeros 5 datos de la combinada
datos_combinados.head()
```

```python
#Filtramos los datos para nuestro Estado Virginia
datos_virginia = datos_combinados[ datos_combinados['state_name'] == 'Virginia']
#Mostramos los primeros 5 datos de la tabla para Virginia
datos_virginia.head()
```

```python
#Exportamos los datos a un archivo en Excel
datos_virginia.to_excel('datos_tarea_no_3.xlsx', index=False)
```
