```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pagina_web='https://es.wikipedia.org/wiki/Premier_League_2023-2024'
```

```python
datos_pl=pd.read_html(pagina_web)
```

```python
datos_pl[10]
```

```python
puntos = datos_pl[10]
puntos['Pts.'].hist()
```

```python
puntos = datos_pl[10]['Pts.']

plt.figure(figsize=(10,6))
sns.histplot(puntos, bins=10, kde=True, color='#FF4500', edgecolor='black', linewidth=1.5, alpha=0.85)

plt.xlabel('Puntos')
plt.ylabel('Frecuencia de Equipos')
plt.title('Distribución de Puntos en la Premier League 2023-2024')
plt.show()
```

```python
puntos.plot.area()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="Pts.", y="Equipo", data=datos_pl[10], palette="Blues_d",hue='Equipo', legend=False)
plt.title('Comparacion de puntos')
plt.xlabel('Puntos')
plt.ylabel('Equipos')
```

```python
goles_favor = datos_pl[10]['GF']
goles_contra = datos_pl[10]['GC']
Equipos=datos_pl[10]['Equipo']
plt.figure(figsize=(10,6))
sns.scatterplot(x=goles_favor, y=goles_contra, hue=Equipos, palette='viridis')
plt.xlabel('Goles a Favor')
plt.ylabel('Goles en Contra')
plt.title('Relación entre Goles a Favor y en Contra por Equipo')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
goles_favor.plot.area()
```

```python
goles=datos_pl[10][['GF','GC']]

goles.plot.hexbin(x='GF',y='GC',gridsize=14)
```

```python
numericas = datos_pl[10][['Pts.', 'G', 'E', 'P', 'GF', 'GC']]
sns.heatmap(numericas)
```

```python
sns.scatterplot(x='Dif.', y='Pts.', data=datos_pl[10], hue='Equipo',s=100)
plt.title('Relación entre Diferencia de Goles y Puntos')
plt.xlabel('Diferencia de Goles')
plt.ylabel('Puntos')
plt.legend(loc='upper left', bbox_to_anchor=(1.05, 1))
```
