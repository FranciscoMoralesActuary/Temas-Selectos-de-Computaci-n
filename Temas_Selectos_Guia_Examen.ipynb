## Supuestos de Regresion Lineal

1.-Colinealidad (Las variables independientes no pueden estar correlacionadas entre si)

2.-Homocedasticidad ($\sigma^2$ es constante)  

3.-Independencia de Errores ($\varepsilon$ ~ v.a.i.i.d)

4.-Linealidad (La relacion entre la variable dependientes ($y$) y las variables independientes ($x_i$) debe ser lineal)

5.-Los errores deben seguir una distribucion normal ( ε  ~  N($\mu ,\sigma^2$) )

6.-No autocorrelacion (Los errores no deben estar correlacionados entre periodos de tiempo)

7.-X debe ser de rango completo (La matriz debe ser de ranfo $n$ donde $n$ es la dimension de la matriz X)

## Minimos Cuadrados Ordinarios

Consiste en hallar los valores $\beta_0$ y $\beta_1$ que hacen
mínima la suma de los cuadrados de las desviaciones entre los valores observados de la variable dependiente $y_i$, es decir

$$
\sum_{i=1}^{n}(y_i-\hat{y_i})^2
$$

En términos más simples, MCO busca la línea que mejor se ajusta a los datos, minimizando el error cuadrático medio

## Multicolinealidad

Ocurre cuando 2 variables predictoras estan fuertemente correlacionadas entre si, esto implica que la información que explican es redundante esto se soluciona eliminando las variables redundantes

Errores que genera la multicolinealidad:

1.-Coeficientes de regresion incorrectos e inestables


2.-Aumento de la varianza de los coeficientes estimados.


3.-Pérdida de precisión en las predicciones (Los intervalos de confianza son incorrectos)


Ilustracion: Si en un modelo de regresión  ingreso y gasto mensual, estan stan altamente correlacionada entre si


multicolinealidad.png

## Heterocedasticidad

Se refiere a que los errores de nuestro modelo tienen varianza constante, cuando esto no ocurre las estimaciones de los coeficientes son imprecisas y los intervalos de confianza incorrectos, esto se soluciona aplicando una transformación a nuestras variable

Causas

1.-Relación no lineal entre variables independientes y dependiente.

2.-Datos no transformados adecuadamente (escalas incorrectas).

3.-Omisión de variables relevantes

Tratamiento

1.-Aplicar transformaciones (logarítmica, raíz cuadrada).

2.-Utilizar regresión generalizada (mínimos cuadrados ponderado  o otro tipo de generalizacion que evite este error).

3.-Weighted Least Squares (WLS): Asignar pesos a las observaciones para dar más importancia a aquellas con menor varianza.

# 10.-Regresión Logística

## Funcion Logit

$\pi=$Probablidad de exito

ODDS = probabilidad de éxito dividida por la probabilidad de fracaso

$$
In\left(\dfrac{\pi}{1-\pi}\right)=\beta_0+\beta_1x
$$

$$
\pi=\dfrac{e^{\beta_0+\beta_1x}}{1+e^{\beta_0+\beta_1x}}
$$

Mediante el uso de ODDS la fucnion logit predice la ocurrencia de un evento entere 0 y 1 donde $\pi$ es nuestra probababilidad de exito

**Parametros Beta** Los paramateros indican el efecto de las variables independientees sobre el logaritmo de los odds

**Interpretacion de las Estimaciones:** Parametros positivos indican que si la variable independiente aunmenta ($x_i$) entonces $p(\pi)$ tambien aumenta y del mismo modo parametros negativos indican que si la variable independiente aunmenta ($x_i$) entonces $p(\pi)$ disminuye

**Negocios**


*   **Marketing** : $P(\pi)=$ Probabilidad de que un cliente realice una compra
*   **Riesgo Crediticio**: $P(\pi)=$ Probabilidad de que un cliente pague su deuda a tiempo
*   **Salud**: $P(\pi)=$ Probabilidad de que un paciente resulte infectado de un virus





## Matriz de Confusion

Metricas



*   **Presicion (Accuracy)**: Da el % de Prediccion Corretas
$$
P=\dfrac{VP+VN}{VP+FP+VN+FN}
$$
*   **Sensibilidad(Recall)**: Que tan bien detecta casos positivos
$$
S=\dfrac{VP}{VP+FN}
$$
*  **Especificidad**: Que tan bien detecta casos negativos
$$
E=\dfrac{VN}{VN+FP}
$$


*   **Precisión Positiva (PPV)**: Da el % de Prediccion de casos positivos, predichos de manera correcta
$$
PPV=\dfrac{VP}{VP+FP}
$$

*   **F1-Score**: Metrica entre presicion y sensibilidad
$$
F_1=\dfrac{2 PPV \cdot S}{PPV + S}
$$
