##Regresin Logstica

```python
# Dividimos el conjunto de datos en dos partes:
# x: variables independientes, eliminando la columna 'Default' del DataFrame
# y: variable dependiente 'Default' para poderla predecir en el modelo
x= datos_entrenar.drop('Default', axis=1)
y= datos_entrenar['Default']
```

```python
# Importamos 'train_test_split' que divide el conjunto de datos en dos partes:
# Conjunto de entrenamiento y el conjunto de prueba para evaluar el modelo
from  sklearn.model_selection import train_test_split
x_entrenar, x_test, y_entrenar, y_test = train_test_split(x, y, test_size=0.2, random_state=42)           # El 20% de los datos los usaremos como conjunto de prueba, mientras que el 80% restante se destinar al entrenamiento
                                                                                                          # random_state=42 asegura que los datos se dividan de la misma manera para garantizar consistencia
```

```python
# Entrena el modelo con el conjunto de entrenamiento ('x_entrenar' y 'y_entrenar')
# Esto permite al modelo entender la relacin entre las variables predictoras y la variable objetivo
# Utiliza un algoritmo para encontrar la mejor linea recta que se ajuste a los datos.
modelo= LogisticRegression()
modelo.fit(x_entrenar, y_entrenar)
```

```python
# Utilizamos "predict" para realizar predicciones sobre un conjunto de datos desconocidos,
# es decir nuestro conjunto de datos de prueba.
predicciones = modelo.predict(x_test)
```

```python
# Importamos 'confusion_matrix', que permite calcular la matriz de confusin del modelo.
# Evala el desempeo del modelo comparando las etiquetas reales ('y_test') con las predicciones ('predicciones')
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predicciones)
```

```python
# Calculamos la precisin del modelo utilizando la funcin 'accuracy_score'
accuracy_score(y_test, predicciones)
```

```python
# Importamos la funcin 'classification_report'  que da un informe detallado de las mtricas de evaluacin del modelo
# Ayuda a evaluar el desempeo del modelo en mayor detalle
from sklearn.metrics import classification_report
print(classification_report(y_test, predicciones))
```

```python
# Calculamos la matriz de confusin y compara el 'y_test' con las predicciones del modelo
# Permite evaluar cmo el modelo clasifica a los clientes que incumplirn el pago y los que no
# Adems convertimos la matriz de confusin en un DataFrame de Pandas
pd.DataFrame(confusion_matrix(y_test, predicciones), columns = ["Predicciones No", "Predicciones Si"], index = ["Real: No", "Real: Si"])

#Notamos que nos arroja un resultado muy poco probable
```

```python
# Calcula los valores para trazar la curva ROC
# Es til para evaluar el rendimiento del un modelo
# La funcin devolver las tasas de verdaderos positivos (TPR) y falsos positivos (FPR) en diferentes niveles de probabilidad
from sklearn.metrics import roc_curve
```

```python
# 'fpr_lr' representa la tasa de falsos positivos y 'tpr_lr' es la tasa de verdaderos positivos calculados
fpr_lr, tpr_lr, _ = roc_curve(y_test, predicciones)

# Esta grfica muestra cmo el modelo clasifica los clientes entre los que incumplen el pago y los que no variando el umbral de decisin
# Visualizamos la curva ROC notando que en efecto como lo percibimos anterioremente, los resultados no estaban siendo muy coherentes
plt.figure(1)
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr_lr, tpr_lr, label='LR')
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('ROC curve LOGISTIC REGRESSION')
plt.legend(loc='best')
plt.show()
print(fpr_lr, tpr_lr)
```

```python
# Para mejorar y acercarnos a resultados cohorentes haremos lo siguiente:
# Eliminaremos el 94% de las filas donde Default=0
fraction_to_remove = 0.94

# Seleccionamos las filas donde 'Default'=0
rows_with_zero = datos_entrenar[datos_entrenar['Default'] == 0]

# Con 'sample' seleccionamos aleatoriamente el 94% de las filas filtradas
rows_to_remove = rows_with_zero.sample(frac=fraction_to_remove, random_state=42).index

# Creamos un nuevo DataFrame eliminando las filas seleccionadas previamente
datos_reducidos = datos_entrenar.drop(index=rows_to_remove)

# Mostramos las primeras 5 filas del DataFrame original
print("DataFrame original:")
datos_entrenar.head(5)
```
