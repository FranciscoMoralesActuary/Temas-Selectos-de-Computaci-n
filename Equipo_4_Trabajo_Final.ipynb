Contamos con las siguientes variables:

1. ID: ID de solicitud de préstamo del cliente
2. Client_Income: Ingresos del cliente en \$
3. Car_Owned: Cualquier automóvil que poseía el cliente antes de solicitar el préstamo para otro automóvil (0 significa No y 1 significa lo contrario)
4. Bike_Owned: Cualquier bicicleta que poseía el cliente (0 significa No y 1 significa lo contrario)
5. Active_Loan: Cualquier otro préstamo activo en el momento de la solicitud del préstamo (0 significa No y 1 significa lo contrario)
6. House_Own: Cualquier casa que poseía el cliente (0 significa No y 1 significa lo contrario)
7. Child_Count: Número de hijos que tiene el cliente
8. Credit_Amount: Monto del crédito del préstamo en \$
9. Loan_Annuity: Anualidad del préstamo en \$
10. Accompany_Client: Quién acompañó al cliente cuando solicitó el préstamo
11. Client_Income_Type: Tipo de ingresos del cliente
12. Client_Education: Nivel más alto de educación alcanzado por el cliente
13. Client_Marital_Status: Estado civil del cliente (D- Divorciado, S- Soltero, M- Casado, W- Viudo)
14. Client_Gender: Género del cliente
15. Loan_Contract_Type: Tipo de préstamo (CL- Préstamo en efectivo, RL- Préstamo renovable)
16. Client_Housing_Type: Situación de vivienda del cliente
17. Population_Region_Relative: Población relativa de la región donde vive el cliente. Un valor más alto significa que el cliente vive en un área más poblada
18. Age_Days: Edad del cliente al momento de enviar la solicitud
19. Employed_Days: Días antes de la solicitud, el cliente comenzó a ganar
20. Registration_Days: Días antes de la solicitud del préstamo, el cliente cambió su registro
21. ID_Days: Días antes de la solicitud del préstamo, el cliente cambió su documento de identidad con el que solicitó el préstamo
22. Own_House_Age: Antigüedad de la casa del cliente en años
23. Mobile_Tag: Número de móvil proporcionado por el cliente (1 significa Sí y 0 significa No)
24. Homephone_Tag: Número de teléfono de casa proporcionado por el cliente (1 significa Sí y 0 significa No)
25. Workphone_Working: ¿Se podía acceder al número de teléfono del trabajo? (1 significa Sí y 0 significa No)
26. Client_Occupation: Tipo de ocupación del cliente
27. Client_Family_Members: Número de familiares que tiene el cliente
28. Cleint_City_Rating: calificación de la ciudad del cliente. 3 indica la mejor, 2 indica buena y 1 indica promedio
29. Application_Process_Day: día de la semana en el que el cliente solicitó el préstamo (0-domingo, 1-lunes, 2-martes, 3-miércoles, 4-jueves, 5-viernes, 6-sábado)
30. Application_Process_Hour: hora del día en el que el cliente solicitó el préstamo
31. Client_Permanent_Match_Tag: indicación si la dirección de contacto del cliente no coincide con la dirección permanente.
32. Client_Contact_Work_Tag: indicación si la dirección de trabajo del cliente no coincide con la dirección de contacto.
33. Type_Organization: tipo de organización en la que trabaja el cliente
34. Score_Source_1: puntuación obtenida de otra fuente. Esta es una fuente normalizada
35. Score_Source_2: puntuación obtenida de otra fuente. Esta es una fuente normalizada
36. Score_Source_3: Puntuación obtenida de otra fuente. Esta es una fuente normalizada
37. Social_Circle_Default: Cuántos amigos o familiares del cliente no cumplieron con algún pago del préstamo en los últimos 60 días
38. Phone_Change: Cuántos días antes de la solicitud del préstamo, el cliente cambió su teléfono
39. Credit_Bureau: Número total de consultas en el último año
40. Default: 1 significa que el cliente no cumplió con los pagos del préstamo y 0 significa lo contrario

##Regresión Logística

```python
# Dividimos el conjunto de datos en dos partes:
# x: variables independientes, eliminando la columna 'Default' del DataFrame
# y: variable dependiente 'Default' para poderla predecir en el modelo
x= datos_entrenar.drop('Default', axis=1)
y= datos_entrenar['Default']
```

```python
# Importamos 'train_test_split' que divide el conjunto de datos en dos partes:
# Conjunto de entrenamiento y el conjunto de prueba para evaluar el modelo
from  sklearn.model_selection import train_test_split
x_entrenar, x_test, y_entrenar, y_test = train_test_split(x, y, test_size=0.2, random_state=42)           # El 20% de los datos los usaremos como conjunto de prueba, mientras que el 80% restante se destinará al entrenamiento
                                                                                                          # random_state=42 asegura que los datos se dividan de la misma manera para garantizar consistencia
```

```python
# Entrena el modelo con el conjunto de entrenamiento ('x_entrenar' y 'y_entrenar')
# Esto permite al modelo entender la relación entre las variables predictoras y la variable objetivo
# Utiliza un algoritmo para encontrar la mejor linea recta que se ajuste a los datos.
modelo= LogisticRegression()
modelo.fit(x_entrenar, y_entrenar)
```

```python
# Utilizamos "predict" para realizar predicciones sobre un conjunto de datos desconocidos,
# es decir nuestro conjunto de datos de prueba.
predicciones = modelo.predict(x_test)
```

```python
# Importamos 'confusion_matrix', que permite calcular la matriz de confusión del modelo.
# Evalúa el desempeño del modelo comparando las etiquetas reales ('y_test') con las predicciones ('predicciones')
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predicciones)
```

```python
# Calculamos la precisión del modelo utilizando la función 'accuracy_score'
accuracy_score(y_test, predicciones)
```

```python
# Importamos la función 'classification_report'  que da un informe detallado de las métricas de evaluación del modelo
# Ayuda a evaluar el desempeño del modelo en mayor detalle
from sklearn.metrics import classification_report
print(classification_report(y_test, predicciones))
```

```python
# Calculamos la matriz de confusión y compara el 'y_test' con las predicciones del modelo
# Permite evaluar cómo el modelo clasifica a los clientes que incumplirán el pago y los que no
# Además convertimos la matriz de confusión en un DataFrame de Pandas
pd.DataFrame(confusion_matrix(y_test, predicciones), columns = ["Predicciones No", "Predicciones Si"], index = ["Real: No", "Real: Si"])

#Notamos que nos arroja un resultado muy poco probable
```

```python
# Calcula los valores para trazar la curva ROC
# Es útil para evaluar el rendimiento del un modelo
# La función devolverá las tasas de verdaderos positivos (TPR) y falsos positivos (FPR) en diferentes niveles de probabilidad
from sklearn.metrics import roc_curve
```

```python
# 'fpr_lr' representa la tasa de falsos positivos y 'tpr_lr' es la tasa de verdaderos positivos calculados
fpr_lr, tpr_lr, _ = roc_curve(y_test, predicciones)

# Esta gráfica muestra cómo el modelo clasifica los clientes entre los que incumplen el pago y los que no variando el umbral de decisión
# Visualizamos la curva ROC notando que en efecto como lo percibimos anterioremente, los resultados no estaban siendo muy coherentes
plt.figure(1)
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr_lr, tpr_lr, label='LR')
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('ROC curve LOGISTIC REGRESSION')
plt.legend(loc='best')
plt.show()
print(fpr_lr, tpr_lr)
```

```python
# Para mejorar y acercarnos a resultados cohorentes haremos lo siguiente:
# Eliminaremos el 94% de las filas donde Default=0
fraction_to_remove = 0.94

# Seleccionamos las filas donde 'Default'=0
rows_with_zero = datos_entrenar[datos_entrenar['Default'] == 0]

# Con 'sample' seleccionamos aleatoriamente el 94% de las filas filtradas
rows_to_remove = rows_with_zero.sample(frac=fraction_to_remove, random_state=42).index

# Creamos un nuevo DataFrame eliminando las filas seleccionadas previamente
datos_reducidos = datos_entrenar.drop(index=rows_to_remove)

# Mostramos las primeras 5 filas del DataFrame original
print("DataFrame original:")
datos_entrenar.head(5)
```
