```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
```

```python
excel_1 = "/content/drive/MyDrive/Tercer_Tarea.xlsx"
```

```python
datos_excel = pd.read_excel(excel_1)
datos_excel.head()
```

```python
excel_2 = "/content/drive/MyDrive/uscounties.xlsx"
datos_fips = pd.read_excel(excel_2)
datos_fips.head()
```

```python
# Hacemos la interseccion entre los dos excels
datos_con_fips = pd.merge(datos_excel, datos_fips, on='fips')
datos_con_fips.head()
```

```python
# Eliminamos los estados que no son Virginia
datos_con_fips.drop(datos_con_fips[datos_con_fips['state_name'] != "Virginia"].index, inplace = True)
datos_con_fips.head()
```

```python
# Hacemos un excel con el dataframe de los condados del estado de virginia
datos_con_fips.to_excel('datos_con_fips.xlsx', index=False)
```
