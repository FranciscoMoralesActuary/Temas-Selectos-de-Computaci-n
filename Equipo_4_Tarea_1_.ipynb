Crear una lista de números secuenciales con el valor mínimo de 10 y el máximo 39. Convierte esta lista en una lista o matriz de 2 dimensiones (3 filas por 10 columnas). Muestra por pantalla el valor del elemento de la esquina superior derecha de la matriz




```python
# Importamos la librería Numpy
import numpy as np
```

```python
# Solución

# Creamos la lista de valores
# Genera una secuencia de números enteros del 10 al 39 excluyendo el 40
# Los guarda dentro de una lista
lista = [n for n in range(10,40)]

# Convertimos la lista en un array
# Los arrays de Numpy son más eficientes que las listas predeterminadas de python
array_lista = np.array(lista)

# Hacemos un reshape a la matriz para convertilar en una de 3x10
# El método reshape reorganiza los datos del array sin alterar los datos númericos guardados
array_lista = array_lista.reshape(3,10)

# Mostramos el elemento de la esquina superior derecha y con formato
# La parte del array [0,-1], se refiere a que accederemos a la primera fila (0), y a la última columna (-1)
print(f"Elemento de la esquina superior derecha = {array_lista[0,-1]}")
```

```python
# Mostramos la matriz para comprobar el resultado con la función print()
print(array_lista)
```

### **Leer desde HTML**

```python
# Guardamos en la variable "pagina_web", es url que contiene información sobre LaLiga, en un
pagina_la_liga = ('https://fbref.com/es/comps/12/Estadisticas-de-La-Liga')

# Extraemos las tablas de la página web con la función "pd.read_html"
# Cada tabla se guarda en un dataframe diferente y además quedan guardadas en una lista
datos_la_liga = pd.read_html(pagina_la_liga)

# Imprimimos un mensaje que nos indica cuántas tablas de la pagina web se extrajeron
print(f"Se extrajeron {len(datos_la_liga)} tablas.")
```

```python
# Definimos la primer tabla de la página con el nombre de 'dataframe' para trabajar más cómodamente
dataframe = datos_la_liga[0]

# Imprimimos las primeras 5 filas de los datos extraidos
dataframe.head()
```

```python
# Importamos las librerías de Matplot y Seaborn
# Para el manejo de gráficos
import matplotlib.pyplot as plt
import seaborn as sns

# Importamos una función para poder anidar gráficos de seaborn
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

# Para poder graficar en 3D y hacer interactivo
import plotly.graph_objects as go
```

```python
# Extraemos la columna llamada Ptsdel DataFrame
puntos = dataframe['Pts']

# Configuramos el tamaño del Histogrma
plt.figure(figsize=(10,6))
# 'puntos': son los datos de la columna pts que se quieren graficar
# 'bins=10': es el número de intervalos en los que se agruparán los datos
# 'kde=True': trazamos una estimación de la densidad junto al histograma (Kernel Density Estimation)
# 'color=#FF4500': le damos un color anaranjado al histograma
# 'edgecolor=black': le damos un color negro al borde del histograma
# 'linewidth=1.5': es el ancho de las líneas que bordean el histograma
# 'alpha=0.85': es la transparencia del histograma
sns.histplot(puntos, bins=10, kde=True, color='#000090', edgecolor='black', linewidth=1.5, alpha=0.85)

# Etiquetamos el eje X como 'Puntos'
plt.xlabel('Puntos')

# Etiquetamos el eje Y como 'Frecuencia de Equipos'
plt.ylabel('Frecuencia de Equipos')

# Ponemos un título al gráfico
plt.title('Distribución de Puntos en LaLiga 2024-2025 en la jornada #7', fontsize = 15)


# Creamos un gráfico más pequeño en el centro derecho que tendrá una estimación de la función acumulativa
# plt.gca() ("Get Current Axes") hace que el nuevo gráfico se coloque dentro del ya existente
# width el tamaño del ancho del gráfico y height la altura del mismo
# loc inserta en el gráfico más grande el pequeño en la ubicación deseada, en este caso centro derecha
grafica_ch = inset_axes(plt.gca(), width="20%", height="20%", loc='center right')

# Graficamos la función acumulativa con las especificaciones de la variable grafica_ch
sns.histplot(puntos, bins=20, cumulative=True, color='#000080', edgecolor='black', stat='density', alpha=0.75, ax=grafica_ch)

# Etiquetas para el gráfico pequeño
grafica_ch.set_title('Estimación de F', fontsize=10)
grafica_ch.set_xlabel(r'$x$')
grafica_ch.set_ylabel(r'$F(x)$')


# Mostramos ambas gráficas en pantalla
plt.show()
```

Podemos ver que la distribución de puntos en LaLiga puede ser ajustado con un modelo conocido, podría ser una Χ cuadrada o una Gamma, esto viendo el histograma y el histograma suavizado por kernel. Tendríamos que validar el modelo con una Prueba de Bondad de Ajuste.

```python
# Configuramos el tamaño de la grafica
plt.figure(figsize=(10, 6))

# Creamos un gráfico de barras usando la libreria seaborn
# - 'x="Pts"': son los valores del eje X que corresponden a la columna 'Pts' del Dataframe
# - 'y="Equipo"': son los valores del eje Y que  corresponden a la columna 'Equipo'
# - 'data=dataframe': especificamos que los datos provienen de un DataFrame
# - 'palette="Blues_d"':usamos una paleta de colores degradados en azul
# - 'hue="Equipo"': especificamos que cada barra será coloreada de acuerdo al equipo
# - 'legend=False': desactivamos la leyenda, ya que los nombres de los equipos ya están en el eje Y
sns.barplot(x="DG", y="Equipo", data=dataframe, palette="Blues_d", hue='Equipo', legend=False)

# Añadimos un título a la gráfica
plt.title('Comparación de Diferencia de Gol')

# Etiquetamos el eje X como
plt.xlabel('Diferencia de Gol')

# Etiquetamos el eje Y como 'Equipos'
plt.ylabel('Equipos')
```

Aquí si podemos observar una cierta relación entre la diferencia de gol y los puntos, ya que vemos que entre mayor y positiva sea esta, más se encuentran adelante en la tabla, salvo excepciones, podemos intentar hacer una gráfica de 3 dimensiones para observar esto.

```python
# Utilizaremos las variables Diferencia de Gol (DG) y Partidos Ganados(PG) para predecir los Puntos (Pts)

# Importamos la librería para realizar regresión lineal múltiple.
# Por ahora solo la usaremos para encontrar los estimadores
# para ajustar bien nuestro plano de regresión en la gráfica
from sklearn.linear_model import LinearRegression

# Extraemos las variables del dataframe de los datos de LaLiga
X = dataframe[['DG', 'PG']] # Variables predictoras, matriz X
y = dataframe['Pts']        # Variable a predecir, vector y

# Creamos el modelo de regresión lineal y lo entrenamos con los datos extraidos
modelo = LinearRegression()
modelo.fit(X, y)

# Obtener los estimadores del plano de regresión para poder usarlos en la gráfica
a = modelo.coef_[0]
b = modelo.coef_[1]
c = modelo.intercept_
```

```python
# Creamos la figura de Plotly
fig = go.Figure()

# Añadimos a la figura los puntos a gráficar en la misma con ayuda de go.Scatter3d, el cual pondrá
# los puntos en el espacio tridimensional, además personalizamos los colores de los puntos y el tamaño de los mismos en la parte de 'marker'
# Ponemos el modo en que se verán con mode='lines+markers' para ponerlos de una forma un poco inusual
fig.add_trace(go.Scatter3d(x=dataframe['DG'], y=dataframe['PG'], z=dataframe['Pts'],
    mode='lines+markers', marker=dict(size=6, color='darkblue', opacity=0.9), name='Datos LaLiga'))


# Generar la cuadrícula de valores para poder graficar el plano de regresión
x_range = np.linspace(dataframe['DG'].min(), dataframe["DG"].max(), 10)
y_range = np.linspace(dataframe['PG'].min(), dataframe['PG'].max(), 10)
x_grid, y_grid = np.meshgrid(x_range, y_range)

# Estimamos el plano de regresión con los datos anteriormente hallados
# Con ayuda de la expresión general de un plano
z_grid = a * x_grid + b * y_grid + c


# Imprimimos a la gráfica nuestro Plano de Regresión, el cual estará definido como una superficie (go.Surface)
# Le ponemos los datos de la cuadrícula generada anteriormente
# Personalizamos el modo en el que se verán en este caso con colorscale="twilight",
# Definimos la opacidad y finalmente nombramos el plano
fig.add_trace(go.Surface(x=x_grid,y=y_grid,z=z_grid,colorscale='twilight',opacity=0.5,name='Plano de Regresión'))


# Actualizar el diseño y además agregamos leyendas a los ejes de nuestro gráfico para poder apreciar que variables estamos viendo
# Adicionalmente ponemos el título de nuestro gráfico
fig.update_layout(scene=dict(xaxis_title='Diferencia de Gol (DG)', yaxis_title='Partidos Ganados (PG)',zaxis_title='Puntos (Pts)', ),
    title='Gráfica 3D con Plano de Regresión')

# Imprimimos el gráfico
fig.show()
```
