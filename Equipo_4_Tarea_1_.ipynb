Crear una lista de nmeros secuenciales con el valor mnimo de 10 y el mximo 39. Convierte esta lista en una lista o matriz de 2 dimensiones (3 filas por 10 columnas). Muestra por pantalla el valor del elemento de la esquina superior derecha de la matriz




```python
# Importamos la librera Numpy
import numpy as np
```

```python
# Solucin

# Creamos la lista de valores
# Genera una secuencia de nmeros enteros del 10 al 39 excluyendo el 40
# Los guarda dentro de una lista
lista = [n for n in range(10,40)]

# Convertimos la lista en un array
# Los arrays de Numpy son ms eficientes que las listas predeterminadas de python
array_lista = np.array(lista)

# Hacemos un reshape a la matriz para convertilar en una de 3x10
# El mtodo reshape reorganiza los datos del array sin alterar los datos nmericos guardados
array_lista = array_lista.reshape(3,10)

# Mostramos el elemento de la esquina superior derecha y con formato
# La parte del array [0,-1], se refiere a que accederemos a la primera fila (0), y a la ltima columna (-1)
print(f"Elemento de la esquina superior derecha = {array_lista[0,-1]}")
```

```python
# Mostramos la matriz para comprobar el resultado con la funcin print()
print(array_lista)
```

### **Leer desde HTML**

```python
# Guardamos en la variable "pagina_web", es url que contiene informacin sobre LaLiga, en un
pagina_la_liga = ('https://fbref.com/es/comps/12/Estadisticas-de-La-Liga')

# Extraemos las tablas de la pgina web con la funcin "pd.read_html"
# Cada tabla se guarda en un dataframe diferente y adems quedan guardadas en una lista
datos_la_liga = pd.read_html(pagina_la_liga)

# Imprimimos un mensaje que nos indica cuntas tablas de la pagina web se extrajeron
print(f"Se extrajeron {len(datos_la_liga)} tablas.")
```

```python
# Definimos la primer tabla de la pgina con el nombre de 'dataframe' para trabajar ms cmodamente
dataframe = datos_la_liga[0]

# Imprimimos las primeras 5 filas de los datos extraidos
dataframe.head()
```

```python
# Importamos las libreras de Matplot y Seaborn
# Para el manejo de grficos
import matplotlib.pyplot as plt
import seaborn as sns

# Importamos una funcin para poder anidar grficos de seaborn
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

# Para poder graficar en 3D y hacer interactivo
import plotly.graph_objects as go
```

```python
# Extraemos la columna llamada Ptsdel DataFrame
puntos = dataframe['Pts']

# Configuramos el tamao del Histogrma
plt.figure(figsize=(10,6))
# 'puntos': son los datos de la columna pts que se quieren graficar
# 'bins=10': es el nmero de intervalos en los que se agruparn los datos
# 'kde=True': trazamos una estimacin de la densidad junto al histograma (Kernel Density Estimation)
# 'color=#FF4500': le damos un color anaranjado al histograma
# 'edgecolor=black': le damos un color negro al borde del histograma
# 'linewidth=1.5': es el ancho de las lneas que bordean el histograma
# 'alpha=0.85': es la transparencia del histograma
sns.histplot(puntos, bins=10, kde=True, color='#000090', edgecolor='black', linewidth=1.5, alpha=0.85)

# Etiquetamos el eje X como 'Puntos'
plt.xlabel('Puntos')

# Etiquetamos el eje Y como 'Frecuencia de Equipos'
plt.ylabel('Frecuencia de Equipos')

# Ponemos un ttulo al grfico
plt.title('Distribucin de Puntos en LaLiga 2024-2025 en la jornada #7', fontsize = 15)


# Creamos un grfico ms pequeo en el centro derecho que tendr una estimacin de la funcin acumulativa
# plt.gca() ("Get Current Axes") hace que el nuevo grfico se coloque dentro del ya existente
# width el tamao del ancho del grfico y height la altura del mismo
# loc inserta en el grfico ms grande el pequeo en la ubicacin deseada, en este caso centro derecha
grafica_ch = inset_axes(plt.gca(), width="20%", height="20%", loc='center right')

# Graficamos la funcin acumulativa con las especificaciones de la variable grafica_ch
sns.histplot(puntos, bins=20, cumulative=True, color='#000080', edgecolor='black', stat='density', alpha=0.75, ax=grafica_ch)

# Etiquetas para el grfico pequeo
grafica_ch.set_title('Estimacin de F', fontsize=10)
grafica_ch.set_xlabel(r'$x$')
grafica_ch.set_ylabel(r'$F(x)$')


# Mostramos ambas grficas en pantalla
plt.show()
```

Podemos ver que la distribucin de puntos en LaLiga puede ser ajustado con un modelo conocido, podra ser una  cuadrada o una Gamma, esto viendo el histograma y el histograma suavizado por kernel. Tendramos que validar el modelo con una Prueba de Bondad de Ajuste.

Aqu si podemos observar una cierta relacin entre la diferencia de gol y los puntos, ya que vemos que entre mayor y positiva sea esta, ms se encuentran adelante en la tabla, salvo excepciones, podemos intentar hacer una grfica de 3 dimensiones para observar esto.

```python
# Utilizaremos las variables Diferencia de Gol (DG) y Partidos Ganados(PG) para predecir los Puntos (Pts)

# Importamos la librera para realizar regresin lineal mltiple.
# Por ahora solo la usaremos para encontrar los estimadores
# para ajustar bien nuestro plano de regresin en la grfica
from sklearn.linear_model import LinearRegression

# Extraemos las variables del dataframe de los datos de LaLiga
X = dataframe[['DG', 'PG']] # Variables predictoras, matriz X
y = dataframe['Pts']        # Variable a predecir, vector y

# Creamos el modelo de regresin lineal y lo entrenamos con los datos extraidos
modelo = LinearRegression()
modelo.fit(X, y)

# Obtener los estimadores del plano de regresin para poder usarlos en la grfica
a = modelo.coef_[0]
b = modelo.coef_[1]
c = modelo.intercept_
```

```python
# Creamos la figura de Plotly
fig = go.Figure()

# Aadimos a la figura los puntos a grficar en la misma con ayuda de go.Scatter3d, el cual pondr
# los puntos en el espacio tridimensional, adems personalizamos los colores de los puntos y el tamao de los mismos en la parte de 'marker'
# Ponemos el modo en que se vern con mode='lines+markers' para ponerlos de una forma un poco inusual
fig.add_trace(go.Scatter3d(x=dataframe['DG'], y=dataframe['PG'], z=dataframe['Pts'],
    mode='lines+markers', marker=dict(size=6, color='darkblue', opacity=0.9), name='Datos LaLiga'))


# Generar la cuadrcula de valores para poder graficar el plano de regresin
x_range = np.linspace(dataframe['DG'].min(), dataframe["DG"].max(), 10)
y_range = np.linspace(dataframe['PG'].min(), dataframe['PG'].max(), 10)
x_grid, y_grid = np.meshgrid(x_range, y_range)

# Estimamos el plano de regresin con los datos anteriormente hallados
# Con ayuda de la expresin general de un plano
z_grid = a * x_grid + b * y_grid + c


# Imprimimos a la grfica nuestro Plano de Regresin, el cual estar definido como una superficie (go.Surface)
# Le ponemos los datos de la cuadrcula generada anteriormente
# Personalizamos el modo en el que se vern en este caso con colorscale="twilight",
# Definimos la opacidad y finalmente nombramos el plano
fig.add_trace(go.Surface(x=x_grid,y=y_grid,z=z_grid,colorscale='twilight',opacity=0.5,name='Plano de Regresin'))


# Actualizar el diseo y adems agregamos leyendas a los ejes de nuestro grfico para poder apreciar que variables estamos viendo
# Adicionalmente ponemos el ttulo de nuestro grfico
fig.update_layout(scene=dict(xaxis_title='Diferencia de Gol (DG)', yaxis_title='Partidos Ganados (PG)',zaxis_title='Puntos (Pts)', ),
    title='Grfica 3D con Plano de Regresin')

# Imprimimos el grfico
fig.show()
```
